```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default')
```

# Pendulum

```python
g = 9.8
l = 2
d = 1
h = 0.1

def F(x):
    theta, omega = x
    return np.array([theta+h*omega, omega+h*(g/l*np.sin(theta)-d*omega)])
```

```python
x0_list = [np.array([0.1, 0.1]), np.array([2, -3]),  np.array([2*np.pi-0.1, -0.1]), np.array([2*np.pi-2, 3]),  ]
```

```python
xs_list = []
for x0 in x0_list:
    xs = [x0]
    for _ in range(150):
        xs.append(F(xs[-1]))
    xs = np.array(xs)
    xs_list.append(xs)
```

```python
colors = iter(plt.cm.rainbow(np.linspace(0, 0.5, len(x0_list))))

plt.figure(figsize=[10,5])
for xs, c in zip(xs_list, colors):
    plt.plot(xs[:,0], xs[:,1], '.', color=c)
    plt.plot(xs[:,0], xs[:,1], alpha=0.5, color=c)
plt.grid()
```

# Logistic Map

```python
r = 3.75

def F(x):
    return r * x * (1-x)
```

```python
x0_list = [0.1, 0.25, 0.5, 0.9]
```

```python
xs_list = []
for x0 in x0_list:
    xs = [x0]
    for _ in range(40):
        xs.append(F(xs[-1]))
    xs = np.array(xs)
    xs_list.append(xs)
```

```python
colors = iter(plt.cm.rainbow(np.linspace(0, 0.5, len(x0_list))))

fig = plt.figure(figsize=(5,2))
ax = fig.add_subplot(111)
for xs, c in zip(xs_list, colors):
    ax.plot(xs, '.', color=c)
    ax.plot(xs, alpha=0.5, color=c)
ax.grid()
ax.set(frame_on=False)  # New
plt.savefig('logisticr={}.png'.format(r), transparent=True)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
xs=np.linspace(0,1,25)
ys=[F(x) for x in xs]
ax.plot(xs,ys, color='cyan')
ax.plot(xs,xs, '--', color='cyan')

cobweb = False
if cobweb:
    for xs in xs_list:
        pts_x=[]
        pts_y=[]
        pts_x.append(xs[0])
        pts_y.append(0)
        for i in range(len(xs)-1):
            pts_x.append(xs[i])
            pts_y.append(xs[i+1])
            pts_x.append(xs[i+1])
            pts_y.append(xs[i+1])
        plt.plot(pts_x, pts_y, '-', color='grey', alpha=0.5)

ax.grid()
ax.set(frame_on=False)  # New
plt.savefig('logistic-liner={}.png'.format(r), transparent=True)
```

# Nonlinear Second Order

```python
a = 0.9
b = 0.5
c = 1

def F(x):
    p, q = x
    return np.array([a*p, b*q + c*p**2])
```

```python
x0_list = [np.array([-1, -1]), np.array([1, 1]),  np.array([1, -1]), np.array([-1, 1]),  ]
```

```python
xs_list = []
for x0 in x0_list:
    xs = [x0]
    for _ in range(150):
        xs.append(F(xs[-1]))
    xs = np.array(xs)
    xs_list.append(xs)
```

```python
colors = iter(plt.cm.rainbow(np.linspace(0, 0.5, len(x0_list))))

plt.figure(figsize=[10,5])
for xs, c in zip(xs_list, colors):
    plt.plot(xs[:,0], xs[:,1], '.', color=c)
    plt.plot(xs[:,0], xs[:,1], alpha=0.5, color=c)
plt.grid()
```
